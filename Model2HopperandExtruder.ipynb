# Model Code

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

# Assuming df is your dataframe and selected_cols is your list of selected columns
# Forward fill missing values
filtered_df2_article2.fillna(method='ffill', inplace=True)
# Backfill missing values
filtered_df2_article2.fillna(method='bfill', inplace=True)

# Standardize the input features
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(filtered_df2_article2[selected_cols[:-1]])

# Standardize the target variable
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(filtered_df2_article2[selected_cols[-1]].values.reshape(-1, 1))

# Convert the scaled features back to a DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=selected_cols[:-1])

# Convert the scaled target variable back to a DataFrame
y_scaled_df = pd.DataFrame(y_scaled, columns=[selected_cols[-1]])

# Split the standardized data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y_scaled_df, test_size=0.2, random_state=42)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Get the coefficient values and feature names
#coefficients = model.coef_[0]
#feature_names = X_scaled_df.columns.tolist()

# Create a DataFrame to combine feature names and coefficients
#coefficients_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})

# Get the coefficient values
coefficients = model.coef_

# Make predictions for training and testing data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the model accuracy (R-squared score) for training and testing data
train_accuracy = r2_score(y_train, y_train_pred)
test_accuracy = r2_score(y_test, y_test_pred)

# Calculate the mean squared error for training and testing data
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print("Coefficient values:", coefficients)
print("Training accuracy (R-squared score):", train_accuracy)
print("Testing accuracy (R-squared score):", test_accuracy)
print("Training Mean Squared Error (MSE):", train_mse)
print("Testing Mean Squared Error (MSE):", test_mse)

# Plot the predicted values against the actual values for testing data
plt.scatter(y_test, y_test_pred)
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.title("Actual vs. Predicted values (Testing data)")
plt.show()
```
