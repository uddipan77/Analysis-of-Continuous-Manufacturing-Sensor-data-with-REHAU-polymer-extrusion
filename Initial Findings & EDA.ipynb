```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# For example, you can read a CSV file like this:
# df = pd.read_csv('your_dataset.csv')

# Choose five columns to plot
columns_to_plot = ['article_type', 'recipe', 'basematerial', 'timecode']
# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
fig.suptitle('Bar Charts for Unique Element Counts in Different Columns')

# Flatten the axes for easy iteration
axes = axes.flatten()

# Plot bar charts for each selected column
for i, column in enumerate(columns_to_plot):
    ax = axes[i]
    column_counts = df[column].value_counts()
    column_counts.plot(kind='bar', ax=ax, color='skyblue')
    ax.set_title(f'Counts for {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Select the column you want to visualize
selected_column = 'order'

# Define the custom order
custom_order = sorted(df[selected_column].unique(), key=lambda x: int(str(x).split('_')[-1]))

# Create a new DataFrame with a custom order
sorted_df = df[df[selected_column].isin(custom_order)].sort_values(selected_column, key=lambda x: x.map({value: i for i, value in enumerate(custom_order)}))



# Get the value counts and sort by index (ascending order)
value_counts_sorted = sorted_df[selected_column].value_counts().sort_index(ascending=False)

# Plot the bar chart
plt.figure(figsize=(30, 10))
value_counts_sorted.plot(kind='bar', color='skyblue')
plt.title(f'Count of Unique Elements in {selected_column}')
plt.xlabel(selected_column)
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
features = metadata_dataframe[['order', 'recipe', 'basematerial','timecode']]
target = metadata_dataframe['article_type']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Creating and fitting the decision tree model for multiclass classification
decision_tree_model = DecisionTreeClassifier(ccp_alpha=0.01)
decision_tree_model.fit(X_train, y_train)

# Plotting the decision tree
plt.figure(figsize=(30, 20))
plot_tree(decision_tree_model, filled=True, feature_names=features.columns, class_names=[str(i) for i in set(target)])
plt.show()
```

```python
y_pred = decision_tree_model.predict(X_test)

# Evaluating the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
